```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
from collections import Counter
%matplotlib inline
from subprocess import check_output
countries_of_the_world=pd.read_csv('/Users/wyman/DataMining/countries of the world.csv',encoding="windows-1252")
country_list=list(countries_of_the_world['Country'].unique())
population_list=list(countries_of_the_world['Population'].unique())
data=pd.DataFrame({'country_list':country_list,'population_list':population_list})
new_index = (data['population_list'].sort_values(ascending=False)).index.values
sorted_data=data.reindex(new_index)
plt.figure(figsize=(150,100))
sns.barplot(x=sorted_data['country_list'], y=sorted_data['population_list'])
plt.xticks(rotation= 45)
plt.xlabel('Countries')
plt.ylabel('Population')
plt.title('Population of the Countries')
```
